```python
brg = [ ]

user = 0

while user != 7 :
	print('''
	1. menambahkan
	2. menghapus
	3. mengedit
	4. menampilkan
	5. cari barang
	6. cari index
	7. stop
	''')
	user = int(input("masukkan fungsi : "))
	
	if user == 1 : #menambahkan
		while True :
			h = input("masukkan barang :")
			brg.extend(h.split(","))
			
			stop = input("ketikkan y untuk berhenti :")
			if stop == 'y' :
				break
		print("list saat ini", brg)
		
	elif user == 2: #menghapus
		while True :
			t = int(input("masukkan index barang yang di hapus : "))
			if t >= len(brg) :
				print("index tidak ditemukan")
			else :
				print(brg.pop(t), "dihapus")
			
			stop = input("ketikkan y untuk berhenti : ")
			if stop == 'y':
				break
		print("barang saat ini ", brg)
		
	elif user == 3 : #mengedit
		while True:
			t = int(input("masukkan index yang akan diganti : "))
			
			if t >= len(brg) :
				print("index tidak ditemukan")
			else :
				brg [t] = input("masukkan nilai baru") 
			
			stop = input("ketikkan y untuk berhenti : ")
			if stop == 'y':
				break
		print("barang saat ini ", brg)
	
	elif user == 4 : #mengakses
		for i in range(len(brg)):
			print("barang sekarang :",brg[i])
			
	elif user == 5 : #cari barang
		cari = input("masukkan barang yang dicari : ")
		
		if cari in brg:
			print("barang ditemukan")
		else:
			print('barang tidak ditemukan')
	
	elif user == 6 : #cari index
		cari3 = input("masukkan barang yang di cari : " )
			
		if cari3 in brg :
			print(f"{cari3} ditemukan pada index ke- {brg.index(cari3)}")
		else :
			print("index tidak ditemukan")
			
	elif user > 7 :
		print("silahkan periksa kembali inputan anda")
```
